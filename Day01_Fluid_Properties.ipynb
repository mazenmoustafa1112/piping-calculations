```python
# Day 01 Fluid Properties & Cavitation Calculator
# File: 05_Python/Day01_fluid_properties.ipynb

def check_cavitation_advanced(suction_pressure, vapor_pressure):
    if suction_pressure <= vapor_pressure:
        return f"DANGER: Cavitation Risk! Suction ({suction_pressure} kPa) <= Vapor ({vapor_pressure} kPa)"
    else:
        return f"SAFE: No cavitation. Suction ({suction_pressure} kPa) > Vapor ({vapor_pressure} kPa)"

# 1. FLUID DATA
rho_fluid = 997.0          # Density [kg/m3] - Water at 25 C
Pvp_fluid = 3.17           # Vapor pressure [kPa abs]
mu_dynamic = 0.000891      # Dynamic viscosity [Pa.s]

# 2. PIPING DATA 
NPS_inch = 4               # Nominal pipe size [inch]
schedule = "STD"           # Pipe schedule
D_m = 0.10226              # Inside diameter [m] (4" STD)
V = 2.0                    # Flow velocity [m/s]

# 3. SYSTEM DATA 
P_suction_actual = 2.5     # Actual suction pressure [kPa abs] 

# CALCULATIONS 
SG = rho_fluid / 1000.0
nu_kinem = mu_dynamic / rho_fluid  # Kinematic viscosity [m2/s]
Re = rho_fluid * V * D_m / mu_dynamic  # Reynolds number [-]

regime = "Turbulent" if Re > 4000 else "Transitional or Laminar"

# Cavitation_Check
cavitation_status = check_cavitation_advanced(P_suction_actual, Pvp_fluid)

# RESULTS 
print("=" * 60)
print(" Fluid Properties & Cavitation Calculator - Day 01")
print("=" * 60)
print(f" Specific Gravity      : {SG:.4f}")
print(f" Kinematic Viscosity   : {nu_kinem*1e6:.3f} mm2/s (cSt)")
print(f" Reynolds Number       : {Re:,.0f}")
print(f" Flow Regime           : {regime}")
print("-" * 60)
print(f" Vapor Pressure (Pvp)  : {Pvp_fluid:.2f} kPa abs")
print(f" Actual Suction Press  : {P_suction_actual:.2f} kPa abs")
print(f" System Status         : {cavitation_status}")
print("=" * 60)
```
